```python
import pandas as pd

# 파일 리드
dat_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝 데이터 전처리 입문_로드북/chap6/data/energydata.csv', sep=',') 

# 선두에서 5행까지 표시
dat_df.head(10)

# 어떤 가정의 가전제품과 조명 등의 상태 로그
```

```python
# 데이터의 행수, 열수를 확인
print(dat_df.shape)
```

```python
# 데이터 형의 확인
print(dat_df.dtypes)
```

```python
# date를 object형에서 datetime형으로 변환 # 왜 데이터셋에서 Timestamp로 되어 있는데 또 포맷을 쓴것은 그냥 변환이 안되서
dat_df['date'] = pd.to_datetime(dat_df['date'], format='%Y-%m-%d %H:%M:%S')
```

```python
# date열의 데이터형을 확인
print(dat_df['date'].dtypes)
```

```python
# date 첫번째 데이터의 형을 확인
print(type(dat_df['date'][0]))
```

```python
print(dat_df['date'][0]); print(dat_df['date'][1]) # 10분 간격 확인
```

```python
dat_df['date']
```

```python
# 전행과 시간차를 계산
dat_df['dif_min'] = dat_df['date'].diff().dt.total_seconds()/60  # diff() 함수로 전행과의 시간차 계산 total_seconds에 의해 초 단위의 값으로 구해짐. 60으로 나누어서 분 단위로 변화 근데 여기서 질문 차라리 total_minute는 없나?
```

```python
dat_df['dif_min'] # 각각 행간의 시간차이다. 즉 분단위로 변환의 시간(분)으로 변환
```

```python
dat_df['dif_min'].isnull().sum()
```

```python
# 선두요소의 결손을 0으로 메꾼다
dat_df['dif_min'] = dat_df['dif_min'].fillna(0)
```

```python
# 선두에서 5행까지 표시
dat_df['dif_min'].head()
```

```python
# 경과시간(분)의 계산
dat_df['cum_min'] = dat_df['dif_min'].cumsum() # 0 인덱스 행에서 기준으로 얼마나 누적경과시간차이 계산

# 원래의 날짜와 경과시간(분)을 표시
dat_df[['date', 'cum_min']].head()
```

```python
# 연습문제1
# 경과시간(시간)의 계산
dat_df['cum_hour'] = (dat_df['cum_min']/60).round(2).astype(int)

# 1시간 단위의 연결번호를 확인
print(dat_df['cum_hour'].unique())

# 원래의 날짜와 경과시간(시)을 표시元
dat_df[['date', 'cum_min', 'cum_hour']].head(10)
```

```python
print(dat_df['cum_hour'].unique())
```

```python
dat_df[['date', 'cum_min', 'cum_hour']].head(10)
```

#### 데이터의 가시화

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Appliances의 꺽은선 그래프
plt.plot(dat_df['cum_min'], dat_df['Appliances'])

# y축(세로축)의 레이블을 추가
plt.xlabel('cum_min')
plt.ylabel('Appliances')
plt.show()
```

```python
# 연습문제3
# T1~T3의 꺽으선 그래프
plt.plot(dat_df['cum_min'], dat_df['T1'], '-r', label='T1')
plt.plot(dat_df['cum_min'], dat_df['T2'], '-g', label='T2')
plt.plot(dat_df['cum_min'], dat_df['T3'], '-b', label='T3')

# 레이블을 추가
plt.xlabel('cum_min')
plt.ylabel('T')
plt.legend()
plt.show()
```
